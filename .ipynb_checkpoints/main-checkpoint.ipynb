```python
import ray
import gym
from ray import tune
from ray.rllib.algorithms.ppo import PPOConfig
from ray.rllib.algorithms.td3 import TD3Config
from ray.rllib.algorithms.sac import SACConfig
from src.dm_envs import acrobot_swingup



#env = "Acrobot-v1"
#env = "CartPole-v1"
#env = "MountainCarContinuous-v0"
#env = "BipedalWalker-v3"
#env = "LunarLander-v2"
#env = "CarRacing-v1" #Error

tune.register_env("AcrobotSwingup", lambda env_ctx: acrobot_swingup())
env = "AcrobotSwingup"
```

```python
#config = PPOConfig()
config = TD3Config()
#config = SACConfig()

config.environment(
    env = env,
    normalize_actions = True,
)

config.resources(
    num_gpus=1,
)

config.rollouts(
    num_rollout_workers= 15,
)

config.framework(
    framework= "tf2",
)

config.training(
    gamma = 0.999,
    train_batch_size= 1000,
    lr= 1e-4,    
    lr_schedule= [
        [0,1e-4],
        [40000, 1e-4],
        [100000, 1e-5]
    ],
    replay_buffer_config={
        "capacity":100000
    }
)
config.exploration(
    exploration_config={
        "stddev":0.2,
        "random_timesteps": 100000,
        "final_scale" : 1
    }
)


# Train via Ray Tune.
# Note that Ray Tune does not yet support AlgorithmConfig objects, hence
# we need to convert back to old-style config dicts.

stop = {"episodes_total" : 20000}
#stop = {"episode_reward_mean" : 250}

analysis = tune.run(
    "TD3",
    name="my_algos",
    config=config.to_dict(),
    stop = stop,
    checkpoint_at_end = True,
    checkpoint_freq=1,
    keep_checkpoints_num=3,
    checkpoint_score_attr= "episode_reward_mean",
    local_dir = "."
    )


```

```python
# Normal Visualization

gym_env = gym.make(env)

trial = analysis.get_best_logdir("episode_reward_mean", "max")
checkpoint = analysis.get_best_checkpoint(
  trial,
  "training_iteration",
  "max",
)
trainer = config.build()
trainer.restore(checkpoint)

observation = gym_env.reset()
done = False
while not done:
  gym_env.render()
  action = trainer.compute_action(observation)
  observation, reward, done, info = gym_env.step(action)
gym_env.close()
```

```python
# DM Mujoco Visualization

import matplotlib.pyplot as plt
import mediapy as media
import numpy as np

gym_env = acrobot_swingup()
trial = analysis.get_best_logdir("episode_reward_mean", "max")
checkpoint = analysis.get_best_checkpoint(
  trial,
  "training_iteration",
  "max",
)
trainer = config.build()
trainer.restore(checkpoint)

observation = gym_env.reset()
done = False
edges = []
while not done:
  edges.append(gym_env.render())
  #plt.imshow(gym_env.render())
  action = trainer.compute_action(observation)
  observation, reward, done, info = gym_env.step(action)
gym_env.close()
print("Displaying video:")
video = media.resize_video(edges, np.array(np.multiply(edges[0].shape[0:2], 10)))
media.show_video(video, fps=30, codec='gif')
```
